### Задача 2.1

Контрольная работа состоит из пяти вопросов. На каждый вопрос приведено четыре варианта ответа, один из которых правильный. Составьте закон распределения числа правильных ответов при простом угадывании. Найдите $M(X)$, $D(X)$.

#### Решение:

Возможные значения случайной величины (в нашем случае правильных ответов) при простом угадывании при пяти вопросах: $x_1 = 0$, $x_2 = 1$, $x_3 = 2$, $x_4 = 3$, $x_5 = 4$, $x_6 = 5$. В задании пять вопросов, а значит мы имееем 5 независимых испытаний. Вероятность правильно угадать ответ из четырех вопросов равна $p = 0.25$.

Посчитаем вероятность каждого значения:


$P(X = 0) = C_n^k p^k (1 - p)^{n - k} = C_5^0 \cdot 0.25^0 \cdot 0.75^{5-0} = \frac{5!}{0! \: (5-0)!} \cdot 0.25^0 \cdot 0.75^{5-0} = 1 \cdot 1 \cdot 0.2373 = 0.2373.$

$P(X = 1) = C_5^1 \cdot 0.25^1 \cdot 0.75^4 = \frac{5!}{1! \: 4!} \cdot 0.25^1 \cdot 0.75^4 = 5 \cdot 0.25 \cdot 0.3164 = 0.3955.$

$P(X = 2) = C_5^2 \cdot 0.25^2 \cdot 0.75^3 = 0.2636.$

$P(X = 3) = C_5^3 \cdot 0.25^3 \cdot 0.75^2 = 0.0878.$

$P(X = 4) = C_5^4 \cdot 0.25^4 \cdot 0.75^1 = 0.0146.$

$P(X = 5) = C_5^5 \cdot 0.25^5 \cdot 0.75^0 = 0.0009.$

Итак, закон распределения такой случайной величины:

Посчитаем математическое ожидание распределения правильных ответов. Поскольку $n = 5$ и $p = 0.25$, получаем:

$$M(X) = n \cdot p = 5 \cdot 0.25 = 1.25.$$

Результат можно понимать так: в среднем при пяти билетах, в к/х четыре ответа и только один правильный, ожидается  $1.25$ правильных ответа, если выбирать наугад.

Дисперсия такой величины:

$$D(X) = n \cdot p \cdot (1 - p) = 5 \cdot 0.25 \cdot (1 - 0.25) = 0.9375.$$

### Задача 2.2

Вероятность попадания в цель при одном выстреле равна $0.01$. Какова вероятность того, что число попаданий при $200$ выстрелах будет не менее $5$ и не более $10$?

#### Решение:

Вероятность того, что число попаданий будет не менее пяти и не более десяти находится в интервале соответствующих значений попаданий в мишень $x=5$ и $x=10$, при $n = 200$ и $p = 0.01$. 

Число испытаний достаточно велико, а вероятность достаточно мала, поэтому воспользуемся формулой Пуассона для нахождения этих вероятностей. Здесь $a = np = 200 \cdot 0.01 = 2$:

$P(X = 5) = \dfrac{a^{k} \cdot e^{-a}}{k!} = \dfrac{2^{5} \cdot e^{-2}}{5!} \approx 0.036089.$

$P(X = 6) = \dfrac{a^{k} \cdot e^{-a}}{k!} = \dfrac{2^{6} \cdot e^{-2}}{6!} \approx 0.012029.$

$P(X = 7) = \dfrac{a^{k} \cdot e^{-a}}{k!} = \dfrac{2^{7} \cdot e^{-2}}{7!} \approx 0.003437.$

$P(X = 8) = \dfrac{a^{k} \cdot e^{-a}}{k!} = \dfrac{2^{8} \cdot e^{-2}}{8!} \approx 0.003437.$

$P(X = 9) = \dfrac{a^{k} \cdot e^{-a}}{k!} = \dfrac{2^{9} \cdot e^{-2}}{9!} \approx 0.000190.$

$P(X = 10) = \dfrac{a^{k} \cdot e^{-a}}{k!} = \dfrac{2^{10} \cdot e^{-2}}{10!} \approx 0.000038.$




Вычисления по формуле Пуассона:

```python
def puasson(n, p, k):
    a = n * p
    return (a ** k) * (np.exp(-a)) / np.math.factorial(k)
```

```python
for i in range (5, 11):
    print(f'P(X={i}) = {puasson(200, 0.01, i)}')
```

Для нахождения искомой вероятности необходимо просуммировать вероятности попадания в мишень в интервале [5, 10]:

```python
sum = 0
for i in range (5, 11):
    sum += puasson(200, 0.01, i)
sum
```

Проверка по формуле Бернулли:

```python
def bernulli(n, p, k):
    return np.math.factorial(n)/(np.math.factorial(k) * np.math.factorial(n - k)) * p**k * (1 - p)**(n - k)
```

```python
sum = 0
for i in range (5, 11):
    sum += bernulli(200, 0.01, i)
sum
```

Таким образом вероятость того, что будет не менее 5 и не более 10 попаданий равна $0.0526$.

### Задача 2.3

В условиях предыдущей задачи сколько выстрелов нужно сделать чтобы быть уверенным с вероятностью $0.9$, что хотя бы $10$ раз будет совершено попадание?

_Подсказка_. Как посчитать $P(X \geq k)$ для величины $X$, имеющей распределение Пуассона? Здесь предстоит немножко покодить.

#### Решение:

Чтобы найти сколько выстрелов нужно сделать необходимо объединять (суммировать) вероятности попадания 10 и более раз. Возьмем диапазон [10, 50] так как попасть 50 и более раз очень маловероятно и практически не влияет на результат. 

Для поиска вероятности просто переберем все возможные количества выстрелов и просуммируем вероятности попасть в мишень больше 10 раз:

```python
# n - количество искомых выстрелов
n = 0

# Основной цикл для перебора n
for i in range(10000):
    summa = 0
    
    # Внутренний цикл для перебора и суммирования искомого диапозона выстрелов [10, 50]
    for k in range (10, 50):
        summa += puasson(i, 0.01, k)
    # Как только вероятность превысила необходимую то выходим из цикла
    if summa >= 0.9:
        n = i
        break
        
print(f'Количество необходимых выстрелов n = {n}')
print(f'Вероятность попасть 10 раз при {n} выстрелах равна {summa}')
```
