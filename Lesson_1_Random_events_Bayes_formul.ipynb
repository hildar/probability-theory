### Задача 1.2

Группа студентов изучает $10$ различных дисциплин. Сколькими способами можно составить расписание на понедельник, если в этот день должно быть $4$ разных занятия?

Определим формулы комбинаторики:

```python
# Размещения
def arrangements(n, k):
    return np.math.factorial(n) // np.math.factorial(n - k)

# Перестановки
def permutations(n):
    return np.math.factorial(n)

# Сочетания
def combinations(n, k):
    return np.math.factorial(n) // (np.math.factorial(k) * np.math.factorial(n - k))
```

Т.к. по условию задания важно чтобы было 4 разных занятия, но расписание обязательно имеет порядлок, то необходимо посчитать сочетания дисциплин. И т.к. минимум 4 дисциплины, то возьмем k=4:

```python
arrangements(10, 4)
```

### Задача 1.3

Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что два определённых человека будут сидеть рядом?

Вероятность того, что человек сядет на одно любое место 1/7, а вероятность того, что второй человек сядет на любое место тоже 1/7. Чтобы найти вероятность того, что два определенных человека будут сидеть рядом нужно просто перемножить их вероятности.

$$P = \frac{1}{7} \cdot \frac{1}{7} = \frac{1}{49}$$

```python
1 / 49
```

Проверим теорию с помощью статистики:

```python
n = 10**3
a = np.random.randint(1, 8, size=n)
b = np.random.randint(1, 8, size=n)
```

```python
spam = np.array([])

# Проверим возможные комбинации совместных событий
for i in range(1, 8):
    for j in range(1, 8):
        if i != j: # Человек же не может одновременно на двух местах сидеть (если он не человек Шредингера)
            m = sum((a == i) & (b == j))
            spam = np.append(spam, m/n)

# Выведем среднее значение
spam.mean()
```

### Задача 1.6

Допустим, имеется некоторая очень редкая болезнь (поражает $0.1 \%$ населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет $99 \%$ больных этой болезнью и всего лишь в $1 \%$ случаев даёт ложный положительный ответ.

_Вопрос_: вакова вероятность, что вы действительно больны ей?

_Подсказка_: вновь используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.

#### Решение:

По условию задачи мы имеем две гипотезы - $B_1$ (болен) и $B_2$ (не болен).

$P(B_1)=0.001$ - вероятность того, что я поражен этой болезнью.

$P(B_2)=0.999$ - вероятность того, что я __не__ поражен этой болезнью.

$P(A | B_1)=0.99$ - вероятность того, что тест выявил болезнь __при условии__, что я поражен болезнью.

$P(A | B_2)=0.01$ - вероятность того, что тест выявил болезнь __при условии__, что я __не__ поражен болезнью.

$P(A)$ - вероятность того, что тест окажется __положительным__, т.е. полной группы наших гипотез $B_1$ и $B_2$.
$$P(A) = P(B_1) \cdot P(A | B_1) + P(B_2) \cdot P(A | B_2)$$
$P(A) = 0.001 \cdot 0.99 + 0.999 \cdot 0.01 = 0.01098$

$P(B_1\:|\:A)$ - вероятность того, что я действительно болен __при условии__, что тест оказался положительным.
$$P(B_1\:|\:A) = \frac{P(A\:|\:B_1) \cdot P(B_1)}{P(A)}$$

$P(B_1|A) = \dfrac{0.99 \cdot 0.001}{0.01098} = 0.0902$

В итоге получилась очень низкая вероятность того, что я действительно болен этой болезнью. А значит есть смысл пересдать тест в другой лаборатории.
