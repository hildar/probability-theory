### Задача 1.2

Группа студентов изучает $10$ различных дисциплин. Сколькими способами можно составить расписание на понедельник, если в этот день должно быть $4$ разных занятия?

Определим формулы комбинаторики:

```python
# Размещения
def arrangements(n, k):
    return np.math.factorial(n) // np.math.factorial(n - k)

# Перестановки
def permutations(n):
    return np.math.factorial(n)

# Сочетания
def combinations(n, k):
    return np.math.factorial(n) // (np.math.factorial(k) * np.math.factorial(n - k))
```

Т.к. по условию задания важно чтобы было 4 разных занятия, но расписание обязательно имеет порядлок, то необходимо посчитать сочетания дисциплин. И т.к. минимум 4 дисциплины, то возьмем k=4:

```python
arrangements(10, 4)
```

### Задача 1.3

Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что два определённых человека будут сидеть рядом?

__Решение:__

Используем классическое определение вероятности. Общее число исходов равно 7!, поскольку 7 человек могут занять скамейку в любом порядке. Найдем число благоприятных исходов. Можно рассуждать так: левый из наших двух определенных людей может занять любое из 6 мест (кроме крайнего правого, поскольку правее него должен сидеть еще один). При этом этим левым человеком может быть любой из двоих. Итак рассадить этих двоих можно $6 \cdot 2$ способами (другими словамми 2 человека могут занять 6 пар мест, при этом порядок их размещения тоже является отдельным благоприятствующим событием - получается 12 вариантов). При этом оставшиеся 5 человек могут занять оставшиеся 5 мест в произвольном порядке. Таким образом число благоприятных исходов $6 \cdot 2 \cdot 5!$. Итак вероятность нашего события:

$$P = \frac{6 \cdot 2 \cdot 5!}{7!} = \frac{2}{7} $$

```python
6 * 2 * permutations(5) / permutations(7)
```

```python
2 / 7
```

__Проверим теорию с помощью статистики:__

```python
# Получим случайное распределение мест на скамейке
def get_bench(n):
    c = np.array([])
    while c.shape[0] < n:
        egg = np.random.randint(1, n + 1)
        if egg not in c:
            c = np.append(c, egg)
    return c
```

```python
# Создадим массив с n вариантами распределения мест на скамейке
n = 10**4 * 3
d = np.empty((0,7), dtype=float)
for i in range(n):
    d = np.append(d, np.resize(get_bench(7), (1,7)), axis=0)
d
```

```python
# Проверим статистическую вероятность, что два определеных человека сядут рядом
f = [1, 5]
m = 0
for bench in d:
    for i in range(7):
        if bench[i] == f[0]:
            friend_1 = i
        if bench[i] == f[1]:
            friend_2 = i
    if abs(friend_1 - friend_2) == 1:
        m += 1
m / n
```

### Задача 1.4

Из $60$ вопросов, входящих в экзаменационные билеты, студент знает $50$. Какова вероятность того, что среди трёх наугад выбранных вопросов студент знает: а) все? б) два?

__Решение:__

__а)__ Общее число исходов при вытягивании 3 билетов из 60 равно $C_{60}^3 = \dfrac{60!}{3!(60 - 3)!} = \dfrac{60!}{3! \cdot 57!} = \dfrac{60 \cdot 59 \cdot 58}{3!} = \dfrac{60 \cdot 59 \cdot 58}{6}$.

Число исходов, благоприятных для события _"Все билеты знакомые"_, равно $C_{50}^3 = \dfrac{50 \cdot 49 \cdot 48}{6}$.

Итак вероятность первого события $P = \dfrac{m}{n} = \dfrac{C_{50}^3}{C_{60}^3} = \dfrac{50 \cdot 49 \cdot 48}{60 \cdot 59 \cdot 58} = 0.5728 $


```python
(50 * 49 * 48) / (60 * 59 * 58)
```

__б)__ Чтобы посчитать число благоприятных исходов для второго события, надо посчитать число возможностей выбрать два знакомых билета (из 50) и 1 незнакомый билет (из 10). Число таких исходов: $C_{50}^2 \cdot C_{10}^1 = \dfrac{50 \cdot 49}{2} \cdot 10 $.

Итак вероятность второго события $P = \dfrac{m}{n} = \dfrac{C_{50}^2 \cdot C_{10}^1}{C_{60}^3} = \dfrac{\frac{50 \cdot 49 \cdot 10}{2}}{\frac{60 \cdot 59 \cdot 58}{6}} = \dfrac{50 \cdot 49 \cdot 10 \cdot 3}{60 \cdot 59 \cdot 58} = 0.358  $

### Задача 1.5

Бросается игральная кость. Пусть событие $A$ - появление чётного числа, событие $B$ - появление числа больше трёх. Являются ли эти события независимыми?

#### Решение:

По определению события $A$ и $B$ являются независимыми, если $P(A|B) = P(A)$. Это эквивалентно равенству $P(AB) = P(A) \cdot P(B)$.

Событие $AB$ соответствует тому, что при броске игральной кости выпало чётное число больше трёх, т.е. 4 или 6. Вероятность такого события равна $\dfrac{1}{3}$. Вместе с тем вероятности событий $A$ и $B$, очевидно, равны $\dfrac{1}{2}$. Перемножая их, получаем $\dfrac{1}{4}$, что не равно $\dfrac{1}{3}$.

Таким образм доказано, что события $A$ и $B$ __зависимы__.

__Мой вариант:__

Данные события являются зависимыми и совместными (т.е. эти события могут случиться одновременно). При этом вероятность наступления каждого события в отдельности равна $1/2$, а условная вероятность события $P(A|B)$ или $P(B|A)$ равна $2/3$ по формуле Байеса.

### Задача 1.6

Допустим, имеется некоторая очень редкая болезнь (поражает $0.1 \%$ населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет $99 \%$ больных этой болезнью и всего лишь в $1 \%$ случаев даёт ложный положительный ответ.

_Вопрос_: вакова вероятность, что вы действительно больны ей?

_Подсказка_: вновь используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.

#### Решение:

По условию задачи мы имеем две гипотезы - $B_1$ (болен) и $B_2$ (не болен).

$P(B_1)=0.001$ - вероятность того, что я поражен этой болезнью.

$P(B_2)=0.999$ - вероятность того, что я __не__ поражен этой болезнью.

$P(A | B_1)=0.99$ - вероятность того, что тест выявил болезнь __при условии__, что я поражен болезнью.

$P(A | B_2)=0.01$ - вероятность того, что тест выявил болезнь __при условии__, что я __не__ поражен болезнью.

$P(A)$ - вероятность того, что тест окажется __положительным__, т.е. полной группы наших гипотез $B_1$ и $B_2$.
$$P(A) = P(B_1) \cdot P(A | B_1) + P(B_2) \cdot P(A | B_2)$$
$P(A) = 0.001 \cdot 0.99 + 0.999 \cdot 0.01 = 0.01098$

$P(B_1\:|\:A)$ - вероятность того, что я действительно болен __при условии__, что тест оказался положительным.
$$P(B_1\:|\:A) = \frac{P(A\:|\:B_1) \cdot P(B_1)}{P(A)}$$

$P(B_1|A) = \dfrac{0.99 \cdot 0.001}{0.01098} = 0.0902$

В итоге получилась очень низкая вероятность того, что я действительно болен этой болезнью. А значит есть смысл пересдать тест в другой лаборатории.
